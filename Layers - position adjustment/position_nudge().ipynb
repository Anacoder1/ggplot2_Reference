```r
library(ggplot2)
```

```r
df <- data.frame(
  x = c(1, 3, 2, 5),
  y = c("a", "c", "d", "c")
)

ggplot(df, aes(x, y)) +
  geom_point() +
  geom_text(aes(label = y))
```

```r
ggplot(df, aes(x, y)) +
  geom_point() +
  geom_text(aes(label = y),
            position = position_nudge(y = -0.1))
```

```r
# Or, in brief

ggplot(df, aes(x, y)) +
  geom_point() +
  geom_text(aes(label = y), nudge_y = -0.1)
```
