```r
library(ggplot2)
```

```r
df <- data.frame(
  x = c(rnorm(100, 0, 3), rnorm(100, 0, 10)),
  g = gl(2, 100)
)

ggplot(df, aes(x)) +
  stat_ecdf(geom = "step")
```

```r
# Don't go to positive / negative infinity

ggplot(df, aes(x)) +
  stat_ecdf(geom = "step", pad = FALSE)
```

```r
# Multiple ECDFs

ggplot(df, aes(x, colour = g)) +
  stat_ecdf()
```
