```r
library(ggplot2)
```

```r
ggplot(faithful, aes(waiting, eruptions)) +
  geom_point() +
  stat_ellipse()
```

```r
ggplot(faithful, aes(waiting, eruptions, 
                     color = eruptions > 3)) +
  geom_point() +
  stat_ellipse()
```

```r
ggplot(faithful, aes(waiting, eruptions,
                     color = eruptions > 3)) +
  geom_point() +
  stat_ellipse(type = "norm", linetype = 2) +
  stat_ellipse(type = "t")
```

```r
ggplot(faithful, aes(waiting, eruptions,
                     color = eruptions > 3)) +
  geom_point() +
  stat_ellipse(type = "norm", linetype = 2) +
  stat_ellipse(type = "euclid", level = 3) +
  coord_fixed()
```

```r
ggplot(faithful, aes(waiting, eruptions, 
                     fill = eruptions > 3)) +
  stat_ellipse(geom = "polygon")
```
