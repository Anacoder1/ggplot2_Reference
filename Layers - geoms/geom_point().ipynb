```r
library(ggplot2)
```

```r
p <- ggplot(mtcars, aes(wt, mpg))
p + geom_point()
```

```r
# Adding aesthetic mappings

p + geom_point(aes(colour = factor(cyl)))
```

```r
p + geom_point(aes(shape = factor(cyl)))
```

```r
# A 'bubblechart':

p + geom_point(aes(size = qsec))
```

```r
# Setting aesthetics to fixed value

ggplot(mtcars, aes(wt, mpg)) + 
  geom_point(colour = "red", size = 3)
```

```r
# Varying alpha is useful for large datasets

d <- ggplot(diamonds, aes(carat, price))
d + geom_point(alpha = 1/10)
```

```r
d + geom_point(alpha = 1/20)
```

```r
d + geom_point(alpha = 1/100)
```

```r
# For shapes that have a border (like 21), you can colour the
# inside and outside separately. Use the stroke aesthetic to
# modify the width of the border

ggplot(mtcars, aes(wt, mpg)) +
  geom_point(shape = 21, colour = "black", fill = "white",
             size = 5, stroke = 5)
```

```r
# You can create interesting shapes by layering multiple
# points of different sizes

p <- ggplot(mtcars, aes(mpg, wt, shape = factor(cyl)))

p + 
  geom_point(aes(colour = factor(cyl)), size = 4) +
  geom_point(colour = "grey90", size = 1.5)
```

```r
p + 
  geom_point(colour = "black", size = 4.5) +
  geom_point(colour = "pink", size = 4) +
  geom_point(aes(shape = factor(cyl)))
```

```r
# geom_point() warns when missing values have been dropped from
# the dataset and not plotted, you can turn this off by setting
# na.rm = TRUE

mtcars2 <- transform(mtcars, mpg = ifelse(runif(32) < 0.2,
                                          NA, mpg))

ggplot(mtcars2, aes(wt, mpg)) + geom_point()
```

```r
ggplot(mtcars2, aes(wt, mpg)) + geom_point(na.rm = TRUE)
```
