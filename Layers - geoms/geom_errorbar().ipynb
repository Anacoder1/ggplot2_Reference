```r
library(ggplot2)
```

```r
# Creating a simple example dataset

df <- data.frame(
  trt = factor(c(1, 1, 2, 2)),
  resp = c(1, 5, 3, 4),
  group = factor(c(1, 2, 1, 2)),
  upper = c(1.1, 5.3, 3.3, 4.2),
  lower = c(0.8, 4.6, 2.4, 3.6)
)

p <- ggplot(df, aes(trt, resp, colour = group))
p + geom_linerange(aes(ymin = lower, ymax = upper))
```

```r
p + geom_pointrange(aes(ymin = lower, ymax = upper))
```

```r
p + geom_crossbar(aes(ymin = lower, ymax = upper), 
                  width = 0.2)
```

```r
p + geom_errorbar(aes(ymin = lower, ymax = upper), 
                  width = 0.2)
```

```r
# Drawing lines connecting group means

p + 
  geom_line(aes(group = group)) +
  geom_errorbar(aes(ymin = lower, ymax = upper),
                width = 0.2)
```

```r
# To dodge bars and errorbars, manually specify the dodge
# width

p <- ggplot(df, aes(trt, resp, fill = group))

p +
  geom_col(position = "dodge") +
  geom_errorbar(aes(ymin = lower, ymax = upper), 
                position = "dodge", width = 0.25)
```

```r
# Because the bars and errorbars have different widths, 
# we need to specify how wide the objects we are dodging are

dodge <- position_dodge(width = 0.9)

p + 
  geom_col(position = dodge) +
  geom_errorbar(aes(ymin = lower, ymax = upper),
                position = dodge, width = 0.25)
```

```r
# When using geom_errorbar() with position_dodge2(), extra
# padding will be needed between the error bars to keep them
# aligned with the bars.

p + 
  geom_col(position = "dodge2") +
  geom_errorbar(aes(ymin = lower, ymax = upper),
                position = position_dodge2(width = 0.5,
                                           padding = 0.5)
               )
```
