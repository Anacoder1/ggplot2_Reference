```r
library(ggplot2)
```

```r
ggplot(diamonds, aes(carat)) +
  geom_density()
```

```r
ggplot(diamonds, aes(carat)) +
  geom_density(adjust = 1/5)
```

```r
ggplot(diamonds, aes(carat)) +
  geom_density(adjust = 5)
```

```r
ggplot(diamonds, aes(depth, colour = cut)) +
  geom_density() +
  xlim(55, 70)
```

```r
ggplot(diamonds, aes(depth, fill = cut, colour = cut)) +
  geom_density(alpha = 0.1) +
  xlim(55, 70)
```

```r
# Stacked density plots - to create a stacked density plot,
# 'count' (density * n) variable instead of the default density

# Loses marginal densities

ggplot(diamonds, aes(carat, fill = cut)) +
  geom_density(position = "stack")
```

```r
# Preserves marginal densities

ggplot(diamonds, aes(carat, stat(count), fill = cut)) +
  geom_density(position = "stack")
```

```r
# Use position = "fill" to produce a conditional density estimate

ggplot(diamonds, aes(carat, stat(count), fill = cut)) +
  geom_density(position = "fill")
```
