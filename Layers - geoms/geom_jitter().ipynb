```r
library(ggplot2)
```

```r
p <- ggplot(mpg, aes(cyl, hwy))
p + geom_point()
```

```r
p + geom_jitter()
```

```r
# Adding aesthetic mappings

p + geom_jitter(aes(colour = class))
```

```r
# Use smaller width / height, to emphasize categories

ggplot(mpg, aes(cyl, hwy)) + geom_jitter()
```

```r
ggplot(mpg, aes(cyl, hwy)) + geom_jitter(width = 0.25)
```

```r
# Use larger width / height to completely smooth away
# discreteness

ggplot(mpg, aes(cty, hwy)) + geom_jitter()
```

```r
ggplot(mpg, aes(cty, hwy)) + geom_jitter(width = 0.5,
                                         height = 0.5)
```
