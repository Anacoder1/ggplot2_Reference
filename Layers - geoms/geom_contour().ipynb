```r
library(ggplot2)
```

```r
# A basic plot

v <- ggplot(faithfuld, aes(waiting, eruptions, z = density))
v + geom_contour()
```

```r
# Or compute from raw data

ggplot(faithful, aes(waiting, eruptions)) +
  geom_density_2d()
```

```r
# Setting bins creates evenly spaced contours in the range
# of the data

v + geom_contour(bins = 2)
```

```r
v + geom_contour(bins = 10)
```

```r
# Setting binwidth does the same thing, parameterised by
# the distance between colours

v + geom_contour(binwidth = 0.01)
```

```r
v + geom_contour(binwidth = 0.001)
```

```r
# Trying other parameters

v + geom_contour(aes(colour = stat(level)))
```

```r
v + geom_contour(colour = "red")
```

```r
v + geom_raster(aes(fill = density)) +
  geom_contour(colour = "white")
```
