```r
library(ggplot2)

# geom_bar() is designed to make it easy to create bar charts
# that show counts (or sums of weights)

g <- ggplot(mpg, aes(class))

# No. of cars in each class:

g + geom_bar()
```

```r
# Total engine displacement of each class

g + geom_bar(aes(weight = displ))
```

```r
# Bar charts are automatically stacked when multiple bars
# are placed at the same location. The order of the fill is
# designed to match the legend

g + geom_bar(aes(fill = drv))
```

```r
# If you need to flip the order (because you've flipped
# the plot), call position_stack() explicitly:

g +
  geom_bar(aes(fill = drv), position = position_stack(reverse = TRUE)) +
  coord_flip() +
  theme(legend.position = "top")
```

```r
# To show (e.g.) means, use geom_col()

df <- data.frame(trt = c("a", "b", "c"), 
                 outcome = c(2.3, 1.9, 3.2))

ggplot(df, aes(trt, outcome)) +
  geom_col()
```

```r
# geom_point() displays the same information & doesn't
# require the y - axis to touch 0

ggplot(df, aes(trt, outcome)) +
  geom_point()
```

```r
# geom_bar() can be used with continuous data, in which case
# it will show counts at unique locations

df <- data.frame(x = rep(c(2.9, 3.1, 4.5), c(5, 10, 4)))

ggplot(df, aes(x)) + geom_bar()
```

```r
# c.f a histogram of the same data

ggplot(df, aes(x)) + geom_histogram(binwidth = 0.5)
```
